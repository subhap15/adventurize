#### Get details of a Venue by passing Venue ID to Detail API

```python
# get venue details
def getvenue_details(venue_id):
    
    df_venuedetail = client.venues(venue_id)
    return (df_venuedetail)
```

```python
def format_details(df_details):
    df_details = df_details.reset_index(drop=True)
    df_group = df_details['attributes'].apply(pd.Series)
    df_group = df_group['groups'].apply(pd.Series)
    df_group.rename(columns={0: "feature_1", 1: "feature_2", 2: "feature_3", 3: "feature_4", 4: "feature_5", 5: "feature_6",6: "feature_7", 7: "feature_8", 8: "feature_9",9: "feature_10", 10: "feature_11", 11: "feature_12"})
    df_moredetails = pd.concat([df_details, df_group], axis=1)
    df_moredetails.rename(columns={0: "feature_1", 1: "feature_2", 2: "feature_3", 3: "feature_4", 4: "feature_5", 5: "feature_6",6: "feature_7", 7: "feature_8", 8: "feature_9",9: "feature_10", 10: "feature_11", 11: "feature_12"})
    
    df_categories = df_moredetails['categories'].apply(pd.Series) 
    df_cat1 = df_categories[0].astype(str).str.strip('{}').str.split(', ', expand=True)
    df_cat1 = df_cat1.drop([4,5,6],axis = 1)
    df_cat1['cat1_id'] = [x.split(':')[1] for x in df_cat1[0]]
    df_cat1['cat1_name'] = [x.split(':')[1] for x in df_cat1[1]]
    df_cat1 = df_cat1.drop([0,1,2,3],axis = 1)
    
    df_cat2 = df_categories[1].astype(str).str.strip('{}').str.split(', ', expand=True)
    df_cat2 = df_cat2.drop([4,5],axis = 1)
    df_cat2['cat2_id'] = [x.split(':')[1] if x != 'nan' else " " for x in df_cat2[0]]
    df_cat2['cat2_name'] = [x.split(':')[1] if x != None else " " for x in df_cat2[1]]
    df_cat2 = df_cat2.drop([0,1,2,3],axis = 1)
    
    df_moredetails = pd.concat([df_moredetails, df_cat1,df_cat2], axis=1)
    df_moredetails = df_moredetails.rename(columns={0: "feature_1", 1: "feature_2", 2: "feature_3", 3: "feature_4", 4: "feature_5", 5: "feature_6",6: "feature_7", 7: "feature_8", 8: "feature_9",9: "feature_10", 10: "feature_11", 11: "feature_12"})
    return df_moredetails
    
```

```python
moredetails = pd.DataFrame()
moredetails = format_details(df_venue_food_details)
```
