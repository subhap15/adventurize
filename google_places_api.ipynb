#### Rating

```python
df_rating = pd.read_csv('venues_clean_bucket.csv', encoding = "ISO-8859-1") 
```

```python
#Extract all rows that have no rating in foursquare
df_rating = df_rating[np.isnan(df_rating['rating'] )]
```

```python
df_rating.columns.values
```

```python
# get rating by calling google places API. Multiple by 2 since google rating is by 5 and foursquare rating is by 10
for i,row in df_rating.iterrows():
    query_result = google_places.nearby_search(
        name = df_rating['name'][i] , keyword=df_rating['name'][i], lat_lng = {'lat': df_rating['lat'][i], 'lng': df_rating['lng'][i]})
    for place in query_result.places:
        place.get_details()
        df_rating['rating'][i] =  place.rating*2 
```

```python
df_rating[df_rating['rating'] == 0]
```

```python
df_rating.to_csv('google_api_rating.csv')
df_rating_bkup = df_rating.copy()
```
