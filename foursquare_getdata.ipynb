#### Get restaurant details by passing venue id

```python
def getvenue_details(venue_id):
    
    df_venuedetail = client.venues(venue_id)
    return (df_venuedetail)
```

```python
df_rest_details = pd.DataFrame()
for i,r in df_restaurant.iterrows(): 
    df_detail = getvenue_details (r['venue.id'])
    df = pd.DataFrame.from_dict(df_detail)
    df = df.transpose()
    df_rest_details = pd.concat([df_rest_details,df])
    
```
