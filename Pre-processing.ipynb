```python
for i, row in df1.iterrows():
    mydict = ast.literal_eval(df1[0][i])
    #print ("mydict  = ", mydict)
    if 'groups' in mydict:
        value = mydict['groups']
        #print ('value', value)
        if not value:
            df1['credit_card'][i] = 'NA'  
            df1['wifi'][i] = 'NA'
            df1['outdoorSeating'][i] = 'NA'
            df1['reservations'][i] = 'NA'
            df1['drinks'][i] = 'NA'
        else:
            for x in range(len(value)):
                #print ('length of list', len(value))
                if value[x]['type'] == 'payments':
                    df1['credit_card'][i] = value[x]['items'][0]['displayValue']
                    #print ('CC', df1['credit_card'][i])
                
                if value[x]['type'] == 'wifi':
                    df1['wifi'][i] = value[x]['items'][0]['displayValue']
                    #print ('CC', df1['wifi'][i])
                    
                if value[x]['type'] == 'outdoorSeating':
                    df1['outdoorSeating'][i] = value[x]['items'][0]['displayValue']
                    #print ('seat', df1['outdoorSeating'][i])
                
                if value[x]['type'] == 'reservations':
                    df1['reservations'][i] = value[x]['items'][0]['displayValue']
                    #print ('reservations', df1['reservations'][i])
                
                if value[x]['type'] == 'drinks':
                    df1['drinks'][i] = value[x]['items'][0]['displayValue']
                    #print ('drinks', df1['drinks'][i])
    else:            
        df1['credit_card'][i] = 'NA'  
        df1['wifi'][i] = 'NA'
        df1['outdoorSeating'][i] = 'NA'
        df1['reservations'][i] = 'NA'
        df1['drinks'][i] = 'NA'
```
